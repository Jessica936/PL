```python
import pandas as pd
from scipy.stats import chi2_contingency
data = pd.read_csv('adult.data.csv', header=None)
data.head()
```

```python
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'sex',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
```

```python
data.head()
```

```python
# 種族是否影響工作收入
cross_table = pd.crosstab(data['race'], data['income'])
print(cross_table)
```

```python
#結婚後身分的改變(或是沒結婚的單身男性女性)是否會對收入造成影響
cross_table2 = pd.crosstab(data['relationship'], data['income'])
print(cross_table2)
```

```python
chi2, p, _, _ = chi2_contingency(cross_table2)
```

```python
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```
