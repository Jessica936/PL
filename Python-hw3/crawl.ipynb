```python
import urllib.request as req
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import json
```

```python
headers={"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36"}
#設置網站
url = 'https://www.1111.com.tw/search/job?d0=140700'
response = requests.get(url)
print (response)
```

```python
soup = BeautifulSoup(response.text, 'lxml')
# 找到所有滿足特定CSS選擇器的元素
jobs = soup.find_all("div",class_="card-body")
print(jobs)
```

```python
for job in jobs:
    if job != None:
        print(job.string)
```

```python
salarys = soup.find_all("div",class_="job_item_detail_salary ml-3 font-weight-style digit_6")
print(salarys)
```

```python
for salary in salarys:
    if salary != None:
        print(salary.string)
```

```python
for job in jobs:
    job_name_element = job.find('h5', class_='card-title title_6')
    job_company_element = job.find('h6', class_='job_item_company mb-1 digit_5 body_3')
    job_pay_element = job.find('div', class_='job_item_detail_salary ml-3 font-weight-style digit_6')
    job_content_element = job.find('p', class_='card-text job_item_description body_4')
    
    # 檢查元素是否存在，以及是否包含文本內容
    job_name = job_name_element.string if job_name_element and job_name_element.string else 'N/A'
    job_company = job_company_element.string if job_company_element and job_company_element.string else 'N/A'
    job_pay = job_pay_element.string if job_pay_element and job_pay_element.string else 'N/A'
    job_content = job_content_element.string if job_content_element and job_content_element.string else 'N/A'
    
    print(" 職缺名稱:", job_name, "\n", "公司名稱:", job_company, "\n", "薪水:", job_pay, "\n", "工作內容:", job_content, "\n", '-'*70)
```

```python
content=[]
for job in jobs:
    job_name_element = job.find('h5', class_='card-title title_6')
    job_company_element = job.find('h6', class_='job_item_company mb-1 digit_5 body_3')
    job_pay_element = job.find('div', class_='job_item_detail_salary ml-3 font-weight-style digit_6')
    job_content_element = job.find('p', class_='card-text job_item_description body_4')
    
    # 檢查元素是否存在，以及是否包含文本內容
    job_name = job_name_element.string if job_name_element and job_name_element.string else 'N/A'
    job_company = job_company_element.string if job_company_element and job_company_element.string else 'N/A'
    job_pay = job_pay_element.string if job_pay_element and job_pay_element.string else 'N/A'
    job_content = job_content_element.string if job_content_element and job_content_element.string else 'N/A'
    
    job_data = {"職缺名稱":job_name, "公司名稱":job_company,"薪水":job_pay,"工作內容":job_content}
    content.append(job_data)
# 資料輸出
df = pd.DataFrame(content)
df.to_csv('1111職缺.csv',encoding='utf-8')
```

```python
fn = "1111職缺.json"
obj = {}
obj["1111職缺"] = []
for job in jobs:
    job_name_element = job.find('h5', class_='card-title title_6')
    job_company_element = job.find('h6', class_='job_item_company mb-1 digit_5 body_3')
    job_pay_element = job.find('div', class_='job_item_detail_salary ml-3 font-weight-style digit_6')
    job_content_element = job.find('p', class_='card-text job_item_description body_4')
    
    # 檢查元素是否存在，以及是否包含文本內容
    job_name = job_name_element.string if job_name_element and job_name_element.string else 'N/A'
    job_company = job_company_element.string if job_company_element and job_company_element.string else 'N/A'
    job_pay = job_pay_element.string if job_pay_element and job_pay_element.string else 'N/A'
    job_content = job_content_element.string if job_content_element and job_content_element.string else 'N/A'
    
    name = {"職缺名稱":job_name, "公司名稱":job_company,"薪水":job_pay,"工作內容":job_content}
    obj["1111職缺"].append(name)
with open(fn, 'w', encoding='utf-8') as json_obj:
    json.dump(obj, json_obj,ensure_ascii=False)
```
